```python
# ライブラリのロード
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
# データのロード
iris = load_iris()
```

```python
pca = PCA(n_components=2)
pca.fit(iris.data)
X = pca.transform(iris.data)
```

```python
# 全データの描画
plt.scatter(X[:,0], X[:,1], alpha=1, c=iris.target)
plt.xlabel("x0")
plt.ylabel("x1")
plt.show()
```
